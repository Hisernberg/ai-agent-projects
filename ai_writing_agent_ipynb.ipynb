```python
from langchain_groq import ChatGroq
from langchain.agents import Tool, initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory

import os
os.environ["GROQ_API_KEY"]=userdata.get('grok')
def summarize_text(text: str) -> str:
    return f"Summarize this: {text}"

def rewrite_text(text: str, tone: str = "professional") -> str:
    return f"Rewrite this in a {tone} tone: {text}"

def blog_intro(topic: str) -> str:
    return f"Write a blog introduction about: {topic}"

# === Define LangChain Tools ===

tools = [
    Tool(
        name="Summarizer",
        func=summarize_text,
        description="Summarizes a piece of text to make it concise."
    ),
    Tool(
        name="Rewriter",
        func=lambda text: rewrite_text(text, "casual"),
        description="Rewrites the given text in a casual tone."
    ),
    Tool(
        name="BlogIntroGenerator",
        func=blog_intro,
        description="Generates an engaging blog introduction for a given topic."
    )
]

# === Load Groq Model ===

llm = ChatGroq(model="mixtral-8x7b-32768")

# === Add Memory ===

memory = ConversationBufferMemory(memory_key="chat_history")

# === Initialize the Agent ===

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION,
    memory=memory,
    verbose=True
)

# === Example Usage ===

query_1 = "Summarize: Artificial Intelligence is rapidly transforming industries like healthcare, finance, and education by automating tasks and uncovering insights from data."
query_2 = "Can you rewrite that in a casual tone?"
query_3 = "Write a blog intro about how AI helps writers."

print(agent.run(query_1))
print(agent.run(query_2))
print(agent.run(query_3))
```

```python
from langchain_groq import ChatGroq
from langchain.agents import Tool, initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory

import os
os.environ["GROQ_API_KEY"]=userdata.get('grok')
def summarize_text(text: str) -> str:
    return f"Summarize this: {text}"

def rewrite_text(text: str, tone: str = "professional") -> str:
    return f"Rewrite this in a {tone} tone: {text}"

def blog_intro(topic: str) -> str:
    return f"Write a blog introduction about: {topic}"


tools = [
    Tool(
        name="Summarizer",
        func=summarize_text,
        description="Summarizes a piece of text to make it concise."
    ),
    Tool(
        name="Rewriter",
        func=lambda text: rewrite_text(text, "casual"),
        description="Rewrites the given text in a casual tone."
    ),
    Tool(
        name="BlogIntroGenerator",
        func=blog_intro,
        description="Generates an engaging blog introduction for a given topic."
    )
]


llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0.7)

# === Add Memory ===

memory = ConversationBufferMemory(memory_key="chat_history")

# === Initialize the Agent ===

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION,
    memory=memory,
    verbose=False
)

# === Example Usage ===

user_query = "Write a blog intro about using AI in creative writing"
print("\n📥 Query:", user_query)
response = agent.run(user_query)
print("\n🤖 Response:", response)

user_query = "Can you rewrite that in a casual tone?"
print("\n📥 Query:", user_query)
response = agent.run(user_query)
print("\n🤖 Response:", response)

user_query= "Can you summarize it ?"
print("\n📥 Query:", user_query)
response = agent.run(user_query)
print("\n🤖 Response:", response)



```
